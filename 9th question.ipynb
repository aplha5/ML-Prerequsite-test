```python
#9.1
import matplotlib.pyplot as mpl
import numpy as np
x=np.arange(0.5,100.5,0.5)
y=x
mpl.plot(x,y)
mpl.xlabel("x")
mpl.ylabel("y")
mpl.title("Plot of y=x for x 0.5 to 100.0 with 0.5 steps")
mpl.show()
```
